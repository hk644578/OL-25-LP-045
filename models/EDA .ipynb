**making three categories male,female,other**

**gender which matches with male assign it Male**

**gender matching with female assign it female**

**assigning others randomly as male and female**

**because others category is very small as compared to others and it will not disturb the variance of male and female in data**

```python
import random
random.seed(42)
def clean_gender(gender):
    gender = str(gender).strip().lower()

    if gender in ['male', 'm', 'cis male', 'man', 'make', 'msle', 'mal', 'maile', 'male-ish', 'malr', 'cis man', 'mail']:
        return 'Male'
    elif gender in ['female', 'f', 'cis female', 'woman', 'femake', 'femail', 'female (cis)', 'trans woman']:
        return 'Female'
    else:
        # Randomly assign either 'Male' or 'Female'
        return random.choice(['Male', 'Female'])
```

```python
df['Cleaned_Gender'] = df['Gender'].map(clean_gender)
```

```python
df.drop(columns=['Gender'], inplace=True)
```

```python
df['Cleaned_Gender'].value_counts().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```
